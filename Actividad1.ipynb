```python
import numpy as np
import easyocr

import matplotlib.pyplot as plt
import cv2 as cv

# Leer imagen original
img = cv.imread("/content/drive/MyDrive/Image detection/placa_q.jpg")
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)

# Detectar negro
lower_black = np.array([0, 0, 0])
upper_black = np.array([180, 255, 50])
mask = cv.inRange(hsv, lower_black, upper_black)

# Filtro de mediana
mask_filtered = cv.medianBlur(mask, 5)

# Guardar máscara
output_path = "/content/drive/MyDrive/Image detection/placaFiltrada.jpg"
cv.imwrite(output_path, mask_filtered)

# Mostrar con matplotlib
plt.imshow(mask_filtered, cmap='gray')
plt.title("Máscara Filtrada")
plt.axis('off')
plt.show()

# OCR con EasyOCR usando CPU
reader = easyocr.Reader(['en'], gpu=False)
results = reader.readtext(output_path)

# Mostrar texto detectado
print("Texto detectado:")
for _, text, conf in results:
    print(f"- {text} (confianza: {conf:.2f})")
```
